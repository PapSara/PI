```python
import math
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

np.random.seed(42)
X = np.random.rand(350, 1)
y = (X + 1.2)**3 - 10 + np.random.randn(350, 1)/10

d = 2

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

tree_reg = DecisionTreeRegressor(max_depth=3, min_samples_leaf=5, max_features=int(math.sqrt(d)), random_state=42)
tree_reg.fit(X_train, y_train)

y_pred = tree_reg.predict(X_test)

mse = mean_squared_error(y_test, y_pred)

def plot_decision_boundary(clf, X, y, axes=[0, 1, -10, 5]):
    x1s = np.linspace(axes[0], axes[1], 100)
    #x2s = np.linspace(axes[2], axes[3], 100)
    #x1, x2 = np.meshgrid(x1s, x2s)
    X_new = x1s.reshape(-1, 1)
    y_pred = clf.predict(X_new)
    custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])

    plt.plot(x1s, y_pred, linewidth=2, label="Decision Boundary")

    plt.scatter(X, y, s=10, c='fuchsia', label="Data Points")

    plt.axis(axes)
    plt.xlabel("Feature 1", fontsize=14)
    plt.ylabel("Target", fontsize=14)
    plt.legend(loc="upper left", fontsize=14)

plt.figure(figsize=(8, 4))
plot_decision_boundary(tree_reg, X, y)
plt.plot([2.45, 2.45], [0, 3], "k-", linewidth=0.5)
plt.plot([2.45, 7.5], [1.75, 1.75], "k--", linewidth=0.5)
plt.plot([4.95, 4.95], [0, 1.75], "k:", linewidth=0.5)
plt.plot([4.85, 4.85], [1.75, 3], "k:", linewidth=0.5)

plt.show()

tree_reg.feature_importances_, mse, tree_reg.predict(X_test)
```
